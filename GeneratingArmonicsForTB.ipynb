```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x= sp.symbols('x')
Tsup = 4e-3
Tmid = 2e-3
Tinf = 0
T = Tsup - Tinf
ft1 = 0.4
ft2 =  1.1
```

```python
n = 10
k=0
a0 = (2/T) *(sp.integrate(ft1*sp.cos(2*sp.pi*k*x/T),(x, -Tinf,Tmid)) + sp.integrate(ft2 * sp.cos(2*sp.pi*k*x/T),(x, Tmid, Tsup)))
fs = a0/2
print("a",k,":", a0/2)
for k in range(1,n):
    ak = (2/T) *(sp.integrate(ft1*sp.cos(2*sp.pi*k*x/T),(x, -Tinf,Tmid)) + sp.integrate(ft2 * sp.cos(2*sp.pi*k*x/T),(x, Tmid, Tsup)))
    bk = (2/T) *(sp.integrate(ft1*sp.sin(2*sp.pi*k*x/T),(x, -Tinf,Tmid)) + sp.integrate(ft2 * sp.sin(2*sp.pi*k*x/T),(x, Tmid, Tsup)))
    print("a",k,":", float(ak))
    print("b",k,":", float(bk))
    fs += ak * sp.cos(2*sp.pi*k*x/T) + bk * sp.sin(2*sp.pi*k*x/T)
```

```python
sp.expand(fs)
```

```python
fs_num = sp.lambdify(x,fs)
xnum = np.arange(Tinf, Tsup, 1e-6)
fs_data = []
for i in xnum:
    fs_data.append(fs_num(i))
```

```python

plt.plot(xnum,fs_data)
```
